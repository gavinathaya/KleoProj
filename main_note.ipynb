## Import packages

```python
import numpy as np
from kleopy.constants import (
    G, m1, m2, ms, l, l1, l2, kappa, T, mu, mu_s)
from kleopy.orbital_eq import (potential, potential_eff)
import kleopy.grid_search as gs
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator
```

## Plot the potentials

```python
%matplotlib widget
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

X = np.arange(-0.6, 0.6, 1e-4)
Y = np.arange(-0.6, 0.6, 1e-4)
X, Y = np.meshgrid(X, Y)
Z1 = potential(X,Y,0)
Z2 = potential_eff(X,Y,0)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z1, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
# Plot the surface.
surf = ax.plot_surface(X, Y, Z2, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
